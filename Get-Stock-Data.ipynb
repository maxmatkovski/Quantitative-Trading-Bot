```python
# testing with 
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# List of Dow Jones Industrial Average stocks
djia_stocks = ['AAPL', 'AXP', 'BA', 'CAT', 'CSCO', 'CVX', 'DIS', 'DOW', 'GS', 'HD', 
               'IBM', 'INTC', 'JNJ', 'JPM', 'KO', 'MCD', 'MMM', 'MRK', 'MSFT', 'NKE', 
               'PG', 'TRV', 'UNH', 'V', 'VZ', 'WBA', 'WMT', 'XOM']

# Setting up the plot
plt.figure(figsize=(15, 10))

# Loop through each stock
for stock in djia_stocks:
    data = yf.download(stock, start="2020-01-01")
    data['TP'] = (data['High'] + data['Low'] + data['Close']) / 3

    # Plotting the Typical Price
    plt.plot(data.index, data['TP'], label=f'{stock}')

plt.title('Typical Price of Dow Jones Stocks (2020 - Present)')
plt.xlabel('Date')
plt.ylabel('Typical Price USD')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import backtrader as bt

# Define the trading strategy
class MovingAverageCrossStrategy(bt.Strategy):
    def __init__(self):
        self.ma_short = bt.ind.SMA(period=10)  # Short-term moving average (10 days)
        self.ma_long = bt.ind.SMA(period=30)   # Long-term moving average (30 days)

    def next(self):
        if self.ma_short[0] > self.ma_long[0] and not self.position:
            self.buy()  # Buy if short MA crosses above long MA
        elif self.ma_short[0] < self.ma_long[0] and self.position:
            self.close()  # Sell if short MA crosses below long MA

def run_backtest(stocks, start_date, end_date):
    for stock in stocks:
        # Download historical data for backtesting
        data = bt.feeds.PandasData(dataname=yf.download(stock, start_date, end_date))

        # Create a cerebro entity
        cerebro = bt.Cerebro()

        # Add strategy to cerebro
        cerebro.addstrategy(MovingAverageCrossStrategy)

        # Set the initial portfolio value
        initial_value = 10000
        cerebro.broker.set_cash(initial_value)

        # Add the data feed
        cerebro.adddata(data)

        # Set the commission
        cerebro.broker.setcommission(commission=0.001)

        # Run the strategy
        cerebro.run()

        # Get final portfolio value
        final_value = cerebro.broker.getvalue()

        # Calculate percentage return
        percentage_return = ((final_value - initial_value) / initial_value) * 100

        # Print out the results
        print(f"Stock: {stock}")
        print(f"Period: {start_date} to {end_date}")
        print(f"Starting Portfolio Value: {initial_value:.2f}")
        print(f"Final Portfolio Value: {final_value:.2f}")
        print(f"Percentage Return: {percentage_return:.4f}%\n")

# Example usage
stocks = ['AAPL', 'MSFT']  # List of stocks
run_backtest(stocks, '2020-01-01', '2021-01-01')
```
